```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(client_credentials_manager = SpotifyClientCredentials(client_id="dcffc14978ed497aa58834a7864a99b7", client_secret="e8c5cc78c8e74ccd92323d146c635b47"))
# Carregar o arquivo CSV
df = data

#Função para obter o ID da música
def get_track_id(track_name, artist_name):
    results = sp.search(q=f'track:{track_name} artist:{artist_name}', type='track', limit=1)
    if results['tracks']['items']:
        return results['tracks']['items'][0]['id']
    else:
        return None
#Função para obter o ID do álbum
def get_album_id(album_name, artist_name):
    results = sp.search(q=f'album:{album_name} artist:{artist_name}', type='album', limit=1)
    if results['albums']['items']:
        return results['albums']['items'][0]['id']
    else:
        return None

# Adicionar coluna para armazenar os IDs de faixas
df['track_id'] = df.apply(lambda row: get_track_id(row['trackName'], row['artistName']), axis=1)

# Adicionar coluna para armazenar os IDs de álbuns
df['album_id'] = df.apply(lambda row: get_album_id(row['albumName'], row['artistName']), axis=1)
```

```python
# Função para obter o ID do artista a partir do nome
def get_artist_id(artistName):
    results = sp.search(q=f'artist:{artistName}', type='artist', limit=1)
    if results['artists']['items']:
        return results['artists']['items'][0]['id']
    else:
        return None

# Adicionar uma nova coluna 'ID do Artista' ao DataFrame
df['ArtistId'] = df['artistName'].apply(get_artist_id)
```
